# Gravtional Waves from GRBs

# 2D graph: mass-frequency Relation

```python
import matplotlib.pyplot as plt
import numpy as np

# Data: Masses (in solar masses) and estimated frequency ranges (Hz)
# X-axis: masses of different astrophysical objects
masses = [1.4, 5, 10, 20, 50]  # Solar masses: typical values for neutron stars, stellar-mass BHs, etc.
labels = ["Neutron Star (1.4)", "Light Black Hole (5)", "Stellar-Mass Black Hole (10)", "Heavy Black Hole (20)", "Hypernova/Collapsar (50)"]

# Y-axis: corresponding gravitational wave frequency ranges (Hz)
frequencies = [500, 200, 100, 30, 1]  # Approximate frequency peaks in Hz

# Plot
plt.figure(figsize=(10, 6))
plt.plot(masses, frequencies, marker='o', linestyle='-', color='b')

# Adding labels for each type of object
for i, label in enumerate(labels):
    plt.text(masses[i], frequencies[i] + 20, label, ha='center', fontsize=10)

# Labels and titles
plt.xlabel("Mass of Object (in Solar Masses)")
plt.ylabel("Peak Gravitational Wave Frequency (Hz)")
plt.title("Approximate Gravitational Wave Frequencies from Different Astrophysical Objects")
plt.yscale('log')  # Log scale for frequency as the range is large

# Customizing ticks and grid for clarity, with x-axis labels rotated
plt.xticks(masses, labels, rotation=45)
plt.grid(True, which="both", linestyle="--", linewidth=0.5)

# Show plot
plt.tight_layout()  # Adjust layout to make room for rotated labels
plt.show()
```

# 3D graph: Relation between the masses $m_1$ and $m_2$ and the GW frequency in a collision

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
c = 3e8  # Speed of light in m/s
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2
M_sun = 1.989e30  # Solar mass in kg

# Mass range (in solar masses) for m1 and m2
m1_range = np.linspace(1, 50, 100)  # Mass of object 1 in solar masses
m2_range = np.linspace(1, 50, 100)  # Mass of object 2 in solar masses

# Meshgrid for m1 and m2 values
M1, M2 = np.meshgrid(m1_range, m2_range)

# Calculate total mass for each (m1, m2) pair and the ISCO frequency in Hz
M_total = (M1 + M2) * M_sun  # Convert to kg
f_ISCO = (c**3) / (6**1.5 * np.pi * G * M_total)  # ISCO frequency formula
# f_ISCO = np.log((c**3) / (6**1.5 * np.pi * G * M_total)) # logarithmic scale

# Set up the 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(M1, M2, f_ISCO, cmap='viridis', edgecolor='none')

# Set the z-axis to logarithmic scale
# ax.set_zscale('log')

# Labels and title
ax.set_xlabel(r"$m_1$ (Solar Masses)")
ax.set_ylabel(r"$m_2$ (Solar Masses)")
ax.set_zlabel("Gravitational Wave Frequency (Hz)")
ax.set_title("Gravitational Wave Frequency as a Function of Binary Masses")

# Add a color bar
fig.colorbar(surf, ax=ax, label="Frequency (Hz)")

# Show plot
plt.show()
```

# Computing the same 3D diagram for wavelength $\lambda = \frac{1}{f_{ISCO}}$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Mass range (in solar masses) for m1 and m2
m1_range = np.linspace(1, 50, 100)  # Mass of object 1 in solar masses
m2_range = np.linspace(1, 50, 100)  # Mass of object 2 in solar masses

# Meshgrid for m1 and m2 values
M1, M2 = np.meshgrid(m1_range, m2_range)

# Constants
c = 3e8  # Speed of light in m/s
G = 6.67430e-11  # Gravitational constant in m^3 kg^-1 s^-2
M_sun = 1.989e30  # Solar mass in kg

# Calculate total mass for each (m1, m2) pair and the ISCO frequency in Hz
M_total = (M1 + M2) * M_sun  # Convert to kg
f_ISCO = (c**3) / (6**1.5 * np.pi * G * M_total)  # ISCO frequency formula
wavelength = 1 / f_ISCO  # Wavelength calculation

# Set up the 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface for gravitational wave wavelength
surf = ax.plot_surface(M1, M2, wavelength, cmap='viridis', edgecolor='none')

# Adding horizontal planes for LIGO-VIRGO and LISA detectability ranges
# LIGO-VIRGO detectability range (approx 10 Hz to 1000 Hz)
z_licho = 0.1  # Lower threshold for LIGO-VIRGO
z_ligo_upper = 1000  # Upper threshold for LIGO-VIRGO

# Create LIGO-VIRGO horizontal plane
X_ligo, Y_ligo = np.meshgrid(m1_range, m2_range)
Z_ligo = np.full(X_ligo.shape, z_licho)
Z_ligo_upper = np.full(X_ligo.shape, z_ligo_upper)
ax.plot_surface(X_ligo, Y_ligo, Z_ligo, alpha=0.3, color='r', label='LIGO-VIRGO Lower Bound')
# ax.plot_surface(X_ligo, Y_ligo, Z_ligo_upper, alpha=0.3, color='r', label='LIGO-VIRGO Upper Bound')

# LISA detectability range (approx 0.1 mHz to 1 Hz)
z_lisa_lower = 0.001  # Lower threshold for LISA
z_lisa_upper = 1  # Upper threshold for LISA

# Create LISA horizontal plane
Z_lisa_lower = np.full(X_ligo.shape, z_lisa_lower)
Z_lisa_upper = np.full(X_ligo.shape, z_lisa_upper)
ax.plot_surface(X_ligo, Y_ligo, Z_lisa_lower, alpha=0.3, color='g', label='LISA Lower Bound')
# ax.plot_surface(X_ligo, Y_ligo, Z_lisa_upper, alpha=0.3, color='g', label='LISA Upper Bound')

# Labels and title
ax.set_xlabel(r"$m_1$ (Solar Masses)")
ax.set_ylabel(r"$m_2$ (Solar Masses)")
ax.set_zlabel("Gravitational Wave Wavelength")
ax.set_title("Gravitational Wave Wavelength as a Function of Binary Masses")

# Add a color bar
fig.colorbar(surf, ax=ax, label="Wavelength (m)")

# Show plot
plt.show()
```
